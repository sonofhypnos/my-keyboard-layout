```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Define keyboard layout
keyboard_layout = [['`',
  '1',
  '2',
  '3',
  '4',
  '5',
  '6',
  '7',
  '8',
  '9',
  '0',
  '-',
  '=',
  'backspace'],
 ['`',
  '1',
  '2',
  '3',
  '4',
  '5',
  '6',
  '7',
  '8',
  '9',
  '0',
  '',
  '`',
  'backspace'],
 ['tab', 'q', 'w', 'e', 'r', 't', 'z', 'u', 'i', 'o', 'p', '',"dead_grave"],
 ['capslock', 'a', 's', 'd', 'f', 'g', 'h', 'j', 'k', 'l', '', "", "ISO_Level3_Shift" 'enter',],
 ['shift', 'y', 'x', 'c', 'v', 'b', 'n', 'm', ',', '.', '-', 'shift'],
 ['ctrl', 'win', 'alt', 'space', 'alt', 'win', 'menu', 'ctrl']]
# Define keybindings with expanded ranges
keybindings = {
    "Win+Space": "Launch Application",
    "Win+Shift+Enter": "Launch Browser",
    "Win+Shift+Space": "Launch Command",
    "Win+Shift+n": "Launch File Browser",
    "Win+Alt+Space": "Launch File Search",
    "Win+n": "Launch Notification Viewer",
    "Win+Enter": "Launch Terminal",
    "Win+Shift+?": "Launch This Dialog",
    "Win+b": "Bluetooth Settings",
    "Win+Alt+0": "Carry Window to Workspace 0",
    "Win+Alt+1": "Carry Window to Workspace 1",
    "Win+Alt+2": "Carry Window to Workspace 2",
    "Win+Alt+3": "Carry Window to Workspace 3",
    "Win+Alt+4": "Carry Window to Workspace 4",
    "Win+Alt+5": "Carry Window to Workspace 5",
    "Win+Alt+6": "Carry Window to Workspace 6",
    "Win+Alt+7": "Carry Window to Workspace 7",
    "Win+Alt+8": "Carry Window to Workspace 8",
    "Win+Alt+9": "Carry Window to Workspace 9",
    "Win+Alt+Ctrl+1": "Carry Window to Workspace 11",
    "Win+Alt+Ctrl+2": "Carry Window to Workspace 12",
    "Win+Alt+Ctrl+3": "Carry Window to Workspace 13",
    "Win+Alt+Ctrl+4": "Carry Window to Workspace 14",
    "Win+Alt+Ctrl+5": "Carry Window to Workspace 15",
    "Win+Alt+Ctrl+6": "Carry Window to Workspace 16",
    "Win+Alt+Ctrl+7": "Carry Window to Workspace 17",
    "Win+Alt+Ctrl+8": "Carry Window to Workspace 18",
    "Win+Alt+Ctrl+9": "Carry Window to Workspace 19",
    "Win+Ctrl+Shift+": "Containing Workspace Up",
    "Win+Ctrl+Shift+": "Containing Workspace Down",
    "Win+Ctrl+Shift+": "Containing Workspace Left",
    "Win+Ctrl+Shift+": "Containing Workspace Right",
    "Win+d": "Display Settings",
    "Win+.": "Load Window Layout",
    "Win+Shift+0": "Move Window to Workspace 0",
    "Win+Shift+1": "Move Window to Workspace 1",
    "Win+Shift+2": "Move Window to Workspace 2",
    "Win+Shift+3": "Move Window to Workspace 3",
    "Win+Shift+4": "Move Window to Workspace 4",
    "Win+Shift+5": "Move Window to Workspace 5",
    "Win+Shift+6": "Move Window to Workspace 6",
    "Win+Shift+7": "Move Window to Workspace 7",
    "Win+Shift+8": "Move Window to Workspace 8",
    "Win+Shift+9": "Move Window to Workspace 9",
    "Win+Ctrl+Shift+1": "Move Window to Workspace 11",
    "Win+Ctrl+Shift+2": "Move Window to Workspace 12",
    "Win+Ctrl+Shift+3": "Move Window to Workspace 13",
    "Win+Ctrl+Shift+4": "Move Window to Workspace 14",
    "Win+Ctrl+Shift+5": "Move Window to Workspace 15",
    "Win+Ctrl+Shift+6": "Move Window to Workspace 16",
    "Win+Ctrl+Shift+7": "Move Window to Workspace 17",
    "Win+Ctrl+Shift+8": "Move Window to Workspace 18",
    "Win+Ctrl+Shift+9": "Move Window to Workspace 19",
    "Win+Ctrl+m": "Move to Scratchpad",
    "Win+Backspace": "Next Window Orientation",
    "Win+,": "Save Window Layout",
    "Win+c": "Settings",
    "Win+Shift+t": "Tile/Float Focus Toggle",
    "Win+i": "Toggle Bar",
    "Win+w": "Wifi Settings",
    "Win+Shift+f": "Window Floating Toggle",
    "Win+f": "Window Fullscreen Toggle",
    "Win+t": "Window Layout Mode",
    "Win+Shift+k": "Window Position Up",
    "Win+Shift+j": "Window Position Down",
    "Win+Shift+": "Window Position Left",
    "Win+Shift+l": "Window Position Right",
    "Win+Shift+": "Window Position Up",
    "Win+Shift+": "Window Position Down",
    "Win+Shift+": "Window Position Left",
    "Win+Shift+": "Window Position Right",
    "Win+Tab": "Next Workspace",
    "Win+Alt+": "Next Workspace",
    "Win+Alt+": "Previous Workspace",
    "Win+Shift+Tab": "Previous Workspace",
    "Win+k": "Relative Window Up",
    "Win+j": "Relative Window Down",
    "Win+": "Relative Window Left",
    "Win+l": "Relative Window Right",
    "Win+": "Relative Window Up",
    "Win+": "Relative Window Down",
    "Win+": "Relative Window Left",
    "Win+": "Relative Window Right",
    "Win+Ctrl+a": "Navigate Scratchpad",
    "Win+Ctrl+Space": "Window by Name",
    "Win+Ctrl+1": "Workspace 11",
    "Win+Ctrl+2": "Workspace 12",
    "Win+Ctrl+3": "Workspace 13",
    "Win+Ctrl+4": "Workspace 14",
    "Win+Ctrl+5": "Workspace 15",
    "Win+Ctrl+6": "Workspace 16",
    "Win+Ctrl+7": "Workspace 17",
    "Win+Ctrl+8": "Workspace 18",
    "Win+Ctrl+9": "Workspace 19",
    "Win+0": "Workspace 0",
    "Win+1": "Workspace 1",
    "Win+2": "Workspace 2",
    "Win+3": "Workspace 3",
    "Win+4": "Workspace 4",
    "Win+5": "Workspace 5",
    "Win+6": "Workspace 6",
    "Win+7": "Workspace 7",
    "Win+8": "Workspace 8",
    "Win+9": "Workspace 9",
    "Win+r": "Enter Resize Mode",
    "Win+Shift+q": "Exit App",
    "Win+Escape": "Lock Screen",
    "Win+Shift+e": "Logout",
    "Win+Shift+p": "Power Down",
    "Win+Shift+b": "Reboot",
    "Win+Shift+r": "Refresh Session",
    "Win+Shift+c": "Reload i3 Config",
    "Win+Ctrl+r": "Restart i3",
    "Win+Shift+s": "Sleep",
    "Win+Alt+q": "Terminate App",
    #Custom
    # Layout Mode
    "Ctrl+Win+Alt+L": "mode Layout Mode: [p]rog [m]ain [l]aptopOnly [d]oublePortrait [o]nlyDoublePortrait [f]rontLaptop [t]rippleLand",
    
    
    # Main Launcher
    "Ctrl+Win+Alt+o": "mode Main Launcher: main gathert[o]wn [f]ocusmate [w]hatsapp [g]mail [k]eep [c]alendar [d]iscord gpt[4] [a]nki [r]eminder [p]rediction [t]odoist",

    
    # Other Shortcuts
    "Win+Shift+Escape": "exec gnome-system-monitor",
    "Win+Alt+Ctrl+q": "exec emacsclient -t",
    "Win+Ctrl+p": "exec xclip -o | sed '/^\s*#/d; s/#.*//' | xclip -selection clipboard",
    "Win+p": "perplexity",
    "Win+e": "[class='Emacs'] scratchpad show",
    "Win+Alt+z": "[class='Zotero'] scratchpad show",
    "Alt+F4": "[con_id='__focused__'] exec --no-startup-id kill -9 $(xdotool getwindowfocus getwindowpid)",
    "Win+BackSpace": "kill",
    "Win+Shift+z": "exec xclip -selection 'c' -out | xargs -I{} xdg-open https://www.google.de/search?q={}",
    "Win+z": "exec echo '' | rofi -dmenu -p 'Search:' | xargs -I{} xdg-open https://www.google.de/search?q={}",
    "Win+w": "claude",
    "Win+q": "gpt",
    "Win+Alt+p": "[class='Google-chrome' instance='fatebook.io'] scratchpad show",
    
    # Mnemonics
    "Ctrl+Win+Alt+r": "mode Layout Mode: [s]enfcall [b]luebutton",
    "s": "mode default exec google-chrome --app=https://public.senfcall.de/ka-rationality",
    "b": "mode default exec google-chrome --app=https://bbb23.senfcall.de/html5client/join?sessionToken=1rjjiv1kgifdse8k",
    
    # Fullscreen toggle
    "Win+Ctrl+f": "exec ~/.screenlayout/doublePortOnly.sh; fullscreen toggle global",

    # Launch Applications
    "Win+c": "exec capture.sh -k 'n'",
    "Win+v": "switch to vertical",
    "Win+a" : "Focus parents",
    #"Win+y": "Focus children", removed
    "Win+y": "copyq menu",
    "Win+g": "split horizontal",
    #For  and similar use XKeysim names. You can find (for  that would be ssharp). You can find them by running "xev" and pressing the corresponding key
    

}

# Define modifier combinations
modifier_combinations = [
    ("Win", "Mod4"),
    ("Win+Shift", "Mod4+Shift"),
    ("Win+Ctrl", "Mod4+Ctrl"),
    ("Win+Alt", "Mod4+Alt"),
    ("Win+Shift+Ctrl", "Mod4+Shift+Ctrl"),
    ("Win+Shift+Alt", "Mod4+Shift+Alt"),
    ("Win+Ctrl+Alt", "Mod4+Ctrl+Alt"),
    ("Win+Shift+Ctrl+Alt", "Mod4+Shift+Ctrl+Alt"),
]

# Define colors
mapped_color = "lightgreen"
unmapped_color = "lightcoral"


# Function to draw keyboard
def draw_keyboard(layout, bindings, modifiers, title):
    fig, ax = plt.subplots()
    x_offset = 0
    y_offset = 0
    key_width = 1
    key_height = 1

    for row in layout:
        x_offset = 0
        for key in row:
            full_binding = f"{modifiers}+{key}" in bindings
            #print(f"{modifiers[0]}+{key}")
            #print(bindings)
            is_modifier = key in modifiers.split("+")
            
            color = mapped_color if (full_binding or is_modifier) else unmapped_color

            rect = Rectangle((x_offset, -y_offset), key_width, key_height, facecolor=color, edgecolor="black")
            ax.add_patch(rect)
            plt.text(x_offset + key_width / 2, -y_offset + key_height / 2, key, ha="center", va="center", fontsize=8)
            x_offset += key_width
        y_offset += key_height

    ax.set_xlim(0, len(layout[0]))
    ax.set_ylim(-len(layout), 0)
    ax.set_aspect('equal')
    ax.axis('off')
    plt.title(title)
    plt.show()

# Draw keyboard for each modifier combination
for mod_name, mod_combination in modifier_combinations:
    #mod_list = mod_combination.split('+')
    #mod_list=[mod_name]
    keybindings = {k.lower():v for k,v in keybindings.items()}
    draw_keyboard(keyboard_layout, keybindings, mod_name.lower(), f"Keybindings: {mod_name}")
```
